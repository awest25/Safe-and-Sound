```python
from transform_csv import rename_string, check_long_lat, transform_long_lat
import numpy as np
import pandas as pd
```

```python
df_crime = pd.read_csv('crime/Crime_Data_from_2020_to_Present_20240420.csv')
```

```python
df_crime.head()
```

```python
# Check where LAT and LONG are 0.0 ie missing
print(df_crime[df_crime["LAT"] == 0.0].shape)
print(df_crime[df_crime["LON"] == 0.0].shape)

# Remove them from the dataset
df_crime = df_crime[df_crime["LAT"] != 0.0]
df_crime = df_crime[df_crime["LON"] != 0.0]
```

```python
# Rename LAT and LON columns to latitude and longitude
df_crime = df_crime.rename(columns={"LAT": "latitude", "LON": "longitude"})
```

```python
df_crime.head()
```

```python
# Continue with usual preprocessing
df_crime = df_crime.rename(columns=rename_string)
df_crime.apply(lambda x: check_long_lat(x["longitude"], x["latitude"]), axis=1)
df_crime["location"] = df_crime.apply(lambda x: transform_long_lat(x["longitude"], x["latitude"]), axis=1)
df_crime = df_crime.drop(columns=["longitude", "latitude"])
```
